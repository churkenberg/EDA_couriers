```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter
```

```python
#analysis of the nationality of couriers and other factors.
#P.S. the sample may not reflect the general population,
#and could also be collected from unscrupulous resources.
```

```python
data = pd.read_excel("/data_delivery 1.xlsx")
```

```python
data
```

```python
plt.boxplot(data['age'])
plt.title('age')
plt.show()
```

```python
#calculating the frequency of gender
#Counting the frequency of each element
word_counts = Counter(data['gender'])

#Printing the frequency table
print("Слово", "\t", "Частота")
for word, count in word_counts.items():
    print(word, "\t", count)
```

```python
#counting the frequency of the race
#Counting the frequency of each element
word_counts = Counter(data['nationality'])

#Printing the frequency table
print("Слово", "\t", "Частота")
for word, count in word_counts.items():
    print(word, "\t", count)
```

```python
plt.hist(data['nationality'], color='skyblue')
plt.ylabel('frequency')
plt.xlabel('nationality')
plt.title('frequency of occurrence of nationalities')
plt.show()
```

```python
sns.displot(data=data["age"], kde=True, color='skyblue')
plt.title('the probability of couriers of a certain age')
plt.show()
```

```python
plt.hist(data['gender'], color='skyblue')
plt.ylabel('frequency')
plt.xlabel('gender')
plt.title('the probability of couriers of a certain gender')
plt.show()
```

```python
print("Descriptive statistics age")
print("Mean: ", np.mean(data['age']))
print("Median: ", np.median(data['age']))
print("Variance: ", np.var(data['age']))
print("Standard deviation: ", np.std(data['age']))
```

```python
#conclusion.
#Couriers are more likely to be men with the nationality of "чурка"
#The average age of couriers is 25 years, and they deviate by an average of 4
```
